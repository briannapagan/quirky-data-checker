````python
from cmr import CollectionQuery, GranuleQuery, ToolQuery, ServiceQuery, VariableQuery
from cmr.queries import CMR_UAT, CMR_OPS
import re
````

Specify your environment and collection short name and version. 

````python

mode = CMR_UAT #CMR_UAT or CMR_OPS for PROD 
short_name = 'M2I3NVAER'
version = '5.12.4'
````

Run ```python_cmr``` granule query

````python
api_granule = GranuleQuery(mode=mode)
api_granule.parameters(
    short_name= short_name,
    version = version,

)
print('number of granules: ' + str(api_granule.hits()))
api_granule_downloadable = api_granule.downloadable()
print('number of downloadable granules: ' + str(api_granule_downloadable.hits()))
````

Select a specific number of random granules to grab via ```.get(some_number)``` or get all granules with ```.get_all()```

````python
#granules = api_granule.get(1000)
granules = api_granule.get_all()
````

Now I will loop over the granules to get a list of opendap URLs

````python
opendap_url_list = []
for i in range(0,np.shape(granules)[0]):
    for element in granules[i]['links']:
        if 'opendap' in element['href']:
            opendap_url_list.append(element['href'])
            break

print('found ' + str(len(opendap_url_list)) + ' opendap links')
````

Now I will do the same loop but only looking for cloud opendap URLs which use the earthdata.nasa.gov in the URL 

````python
cloud_opendap_url_list = []

search_component = 'opendap.*earthdata.nasa.gov'
regex_pattern = re.compile(search_component)


for i in range(0,np.shape(granules)[0]):
    for element in granules[i]['links']:
        if regex_pattern.search(element['href']):
            cloud_opendap_url_list.append(element['href'])
            break

print('found ' + str(len(cloud_opendap_url_list)) + ' cloud opendap links')
````

````python
onprem_opendap_url_list = []
search_component = 'gesdisc.eosdis.nasa.gov.*opendap'
regex_pattern = re.compile(search_component)


for i in range(0,np.shape(granules)[0]):
    for element in granules[i]['links']:
        if regex_pattern.search(element['href']) is not None and '.html' not in element['href']:
            onprem_opendap_url_list.append(element['href'])
            break

print('found ' + str(len(onprem_opendap_url_list)) + ' on-prem opendap links')
````

````python
if len(onprem_opendap_url_list) + len(cloud_opendap_url_list) == len(opendap_url_list):
    print('success! on prem and cloud URLs sum to total number of opendap URLs found')
else:
    print('WARNING: on prem and cloud URLs DO NOT sum to total number of opendap URLs found')
````
